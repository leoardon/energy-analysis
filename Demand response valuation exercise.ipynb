## Preface

This report outlines our group's findings and recommendations to the management of Kaluza, covering the ongoing trial of demand-response (DR) technology on the Orkney Isles and the question whether there is a business case for a commercial implementation.

Our findings and conclusions are based on the exploratory analysis of the wind turbine telemetry and load data provided, as well as on the business insights offered at a Q&A session with stakeholders on September 2nd.

## Executive summary & recommendations

Based on our initial study of the Orkney DR potential, we conclude that the HSO scheme has a potential to be highly beneficial for both local wind energy generators and residents.

Based on our estimate, a 50% DR penetration can reduce households' energy costs by 16% and curtailment energy losses by over 80%.

We believe though that for the project to be commercially viable, the government / utilities' assistance is required for the initial investment in modern storage heaters.

The scheme is worth pursuing further by engaging regional and federal energy and environmental regulators and utilities in order to secure funding for the required upgrade of domestic heating systems on the islands. 

Should such funding be available, we recommend that Kaluza proceeds with the HSO project.

### Definitions

* <b>kW vs kWh</b>: kW is a measure of power while kWh is a measure of energy (to be precise: the amount of energy required to keep a 1,000 watt appliance running for an hour). So a 1,000 watt appliance needs 1,000 watts (1 kW) of power to make it work, and uses 1 kWh of energy in an hour.
* <b>Setpoint</b>: A limit to a turbine power output, set by the grid operator. Dynamic management of the output from all grid-connected generators allows the the grid operator to balance the overall system in response to demand.
* <b>Curtailment:</b> A wind turbine is in curtailment whenever its power output is at the actual Setpoint value (or slightly above, allowing for tolerances). If the power output is natuarally below the Setoint, curtailment is not taking place, because there is no sufficient wind to activate the Setpoint limitations. Similarly, a Setpoint equal to a turbine's nameplate (max) capacity can be referred to as a default setting, when the turbine's generation is unrestricted.  
* <b>Lost energy:</b> We measure the lost energy, expressed in kWh (see above), during a curtailment event by comparing the actual power output with an estimate of the power output that would have been achieved given the actual wind speed.

## Current situation

**Project Background**

Orkney Isles are a set of islands off the North tip of Scotland. The area has a great potential for renewable energy (RE) generation, in particular wind (both on- and off-shore) and tidal energy. Thanks to various energy initiatives and government grants the islands moved to forefront of the green energy movement and are now home to almost 50MW wind generation capacity and a hub for tidal energy technology development. 

With the growth of the RE projects on the islands, Orkney renewable power generation reached 120% of the local electricity consumption by 2013 and the islands became a net power exporter. 

Having said that, the intermittence of the RE generation means that the islands continue to rely on the power imported though the national grid. In 2017 the islands imported electricity 5 months in the year <sup>1</sup>. The 40 MW connector to the mainland UK is used for electricity import and export. 

The existing infrastructure has insufficient capacity for export of all excess power generated on the islands, which results in frequent curtailment of RE generation. This in turn leads to lower revenues and returns from the RE projects(some of which are owned by the local communities). 

In fact, in 2012 the distribution company operating in Orkney (SHEPD) introduced a moratorium on the connection of renewables to the distribution network on the islands <sup>2</sup>, thus restricting new RE projects.

A new £260 million transmission link from the Orkney to the mainland, which will provide an additional 220MW of capacity is planned to be constructed by 2023, subject to sufficient number of new RE projects approved for execution. Nevertheless, until the sufficient export capacity is available, local solutions to the RE oversupply are required. 

*Electricity in energy mix*

Electricity constituted 46% of the total energy consumption by local households in 2017 <sup>3</sup>. The islands are not part of the central gas heating system. Electricity is used for approximately half of the water and space heating, which explains the high level of electricity consumption by the islands' households (and the high utility bills) - 7,596 kWh per household in 2018, double the UK average of 3,799 kWh for the year.

Oil is still used for about a quarter of domestic heating, and further 14% is fueled by solid fuel <sup>4</sup>. 

*Energy poverty*

Remoteness of the location, harsh climate conditions, and the old housing stock translate into one the highest rates of 'energy poverty' in the country. This means that a large share of the islands' population spend more than 10% of their household income on energy.

Orkney isles' residents are at disadvantage from both income and cost side, as their average income is 86% of the UK average of Orkney overall and 65% of the UK average for the remote islands <sup>5</sup> and they are subject to one of the highest electricity tariffs in the country, due to the remoteness of the location and its climate.

**Project Objectives**

The aim of the project is to estimate a potential to reduce curtailment-related energy losses by shifting consumers’ demand for electricity. That shift would be enabled by introduction of storage heaters to Orkney households. The ultimate goal is therefore to determine whether there is a commercially viable scheme of DR that would benefit consumers, generators and the overall communities on Orkney Isles.

### Estimating impact from curtailment

**Turbine power curve**

To quantify the impact of curtailment, out first task is to estimate the power output that would have been achieved from the turbine had there been no restrictions imposed on it.

Reviewing the uncurtailed telemetry data we identify a typical wind turbine power curve, which we separate into 4 distinct segments. This separation is important, as the relationship between power and wind speed is driven by different factors in each of them.

1. <b>Low-wind (0-3 m/s):</b> Power output = 0 kW (wind speed is insufficient for power generation).
2. <b>Ramp-up (3-13 m/s):</b> Power output increasing as a function of wind speed (increasing).
3. <b>Full capacity (13-29 m/s):</b> Power output = 900 kW (max capacity).
4. <b>Safety ramp-down (>29 m/s):</b> Power output decreasing linearly (wind speed is too strong and the wind turbine needs to be slow down mechanically).

### Energy lost due to curtailment (Answer to Q1)

Using our power curve function, we get to an estimate of the energy lost due to curtailment: we calculate the power output that would have been achieved during those periods and deduct the actual power output (constrained by Setpoint).

```python
telemetry_uc = telemetry[~telemetry['curtailed']].copy()
telemetry_uc = telemetry_uc.sort_values(by=["Wind_ms"])

telemetry_uc.loc[
    (telemetry_uc["Wind_ms"] > 3) &
    (telemetry_uc["Wind_ms"] < 29) &
    (telemetry_uc["Power_kw"].abs() <= 1e-5)
] = np.nan
telemetry_uc.dropna(inplace=True)

telemetry_uc_l = telemetry_uc[
    (telemetry_uc["Wind_ms"] > 3) &
    (telemetry_uc["Wind_ms"] < 29)
].copy()

from scipy.optimize import curve_fit

def logistic_func(x, k, x0):    
    return 900 / (1 + np.exp(-k * (x-x0)))

p0 = (1/80, 10)

popt_logis, _ = curve_fit(logistic_func, telemetry_uc_l["Wind_ms"], telemetry_uc_l["Power_kw"], p0)

telemetry_uc_e = telemetry_uc[
    (telemetry_uc["Wind_ms"] >= 29)
]


def exp_func(x, a, b):    
    return a*(b**x)

p0 = (400e6, 1/2)

popt_exp, _ = curve_fit(exp_func, telemetry_uc_e["Wind_ms"], telemetry_uc_e["Power_kw"], p0)

def power_uncurtailed(x):
    return np.where(
        x <= 3, 0, 
        np.where(
            x < 29, logistic_func(x, *popt_logis), 
            exp_func(x, *popt_exp)
        )
    )

y_hat = power_uncurtailed(telemetry_uc["Wind_ms"])



telemetry["Power_potential_kw"] = np.where(telemetry['curtailed'], 
                                           power_uncurtailed(telemetry["Wind_ms"]),
                                           telemetry['Power_kw'])

telemetry['est_loss_kw'] = telemetry['Power_potential_kw'] - telemetry['Power_kw']
m = telemetry["est_loss_kw"] < 0.
telemetry.loc[m, "est_loss_kw"] = 0.

power_loss_df = telemetry.copy().drop(['curtailed'], axis = 1)

dt_min = power_loss_df.index.min().date()
dt_max = power_loss_df.index.max().date() + dt.timedelta(days=1)
period = (dt_max - dt_min).days * 24 * 60

t = power_loss_df.index
r = pd.date_range(dt_min, periods=period, freq='T')

power_loss_df_ri = power_loss_df.reindex(t.union(r)).interpolate('time').loc[r]
power_loss_df_ri.dropna(inplace=True)

energy_hr = power_loss_df_ri.resample("h").mean()
energy_hr.rename(columns={
    'Power_kw':'energy_kwh', \
    'Power_potential_kw':'est_uncurtailed_kwh','est_loss_kw': 'est_loss_kwh'
}, inplace = True)
```

```python
energy_day =energy_hr.resample("d").sum().drop(['Setpoint_kw', 'Wind_ms'], axis = 1)
fig, axs = plt.subplots(2, figsize=(20, 12), sharex=False)

plt.suptitle('Actual Wind Generation vs Estimated Uncurtailed, avg daily kWh', fontsize = 20)

for i, year in enumerate([2016, 2017]):

    energy_produced = energy_day[energy_day.index.year == year]['energy_kwh']
    energy_uncurtailed = energy_day[energy_day.index.year == year]['est_uncurtailed_kwh']

    axs[i].plot(energy_uncurtailed, "r.-", alpha=0.6, label="Estimated Uncurtailed")
    axs[i].plot(energy_produced, label="Actual Generation")
    axs[i].fill_between(
        energy_produced.index, energy_produced, energy_uncurtailed, 
        color="red", alpha=0.3, label="Lost"
    )
    axs[i].legend(loc="upper left")
    #plt.title(f'{year} Actual Wind Generation vs Estimated Uncurtailed, avg daily kWh')
```

## Business case analysis

### Scope & addressable market

The core premise of our analysis is that controlling a share of domenstic electricity demand can achieve material improvements to the energy market on Orkney islands. That would involve reduction in (and ideally elimination of) curatilment of RE generation, as well as reduction in cost of energy to local households. 

Use of modern storage heaters can be a practical and simple way to implement solution to this significant problem.

We start by considering the households which already use electricity for space heating as the base case. These are the people who are likely to feel immediate benefits from the HSO scheme, as they would be shifting from the most expensive source of heat (electricity at peak tariffs).

Based on our research we estimate that 43% of the population of the Orkney isles (10,385 households) uses electricity for heating. For those who do, we know that on average 75% of the total electricity consumption (demand) relates to space heating <sup>9</sup>. These data points allow us to impute 'flexible' consumption which can be shifted in time.

In this context we see the households who already use electrical heating as the 100% of population for the DR purposes, as these are the households who are in the position to immediately change their electricity demand. 

Future analysis could include additional scenarios where we estimate the impact of conversion of non-electricity users (in the context of heating) to electricity.

We consider the efficiency of the HSO scheme from two perspectives: 
(1) that of the power generators interested in reduction in curtailment and improvement of their investment returns, and
(2) that of the local households, seeking to reduce their electricity costs.

### Modelling assumptions

#### Investment cost

To enable the DR-scheme, an initial investment will be required per participating household - consisting of <sup>10</sup>:

* 2 x quantum storage heaters with 7140kWh storage, at a cost of £556 each (cost is assumed to include the expenses associated with installation). The heaters specificatation and number is based on the consumption numbers, so that they would be sufficient for the maximum daily heating-related consumption in a year.
* 1 smart unit, that would control the heater(s), at a cost of £162

The default case assumes that the cost of the hardware (storage heaters and smart units) will be covered by consumers or - ideally - by a government grant. However, we incorporated a flexibility to vary the split of cost burden between consumers and the local generators, participating in the programme. 

#### Additional revenues to generators

We assume that all generators are subject to identical and simultaneous curtailment changes (simplification from the actual 'stacked' access to grid). 

We also assume that all generators receive payments per kWh produced that would be comparable with the lower end of the tariff range paid by consumers. This assumption is important, as it allows to offer discouts on curtailed energy commensurate with consumers default costs. Therefore, under our model all generators are entitled to the same level of Feed-in-Tariff (11p/kWh) and export tariff (3p/kWh).

As the result, the incentive to consumers to participate in the DR scheme is offered though a discount to their electricity tariffs and applied to the incremental energy used in curtailment period (can be done in a form of rebates, based on the DR records). The rebates represent sharing of the generators' incremental revenues with consumers. We tested 3 levels of revenue sharing levels in our analysis: 25%, 50% and 75% of generator's benfit from the scheme. 

An option of sharing cost of heaters is also included.

#### Cost differential to households

Based on the Electricity statistics, there are 2 Orkney HHs use 2 tariffs ('Standard' and 'Economy 7'). 
The Standard tariff is flat (17.56p/kWh +standing charge of 26.53 per day), whereas the Economy 7 offers low rate for 7 hours at night (11pm-7am), with day tariff of 18.71p/kWh and night of 13.15p/kWh +30.93p per day <sup>13</sup>. 

Considering that we are focusing on the households who use electric heating, we assume that they are already on the 'Economy 7' tariff or will convert to it as soon as they install storage heaters and agree to participate in the HSO DR scheme. We additionally confirmed that it would be more beneficial for the average household to be on the E7 tariff, even at the default consumption profile.

Additional discount offered by wind generators as a DR incentive is included in the calculation as well.

#### Savings / Benefit Quantification

Generators: Reduction in annual output losses related to generation curtailment and incremental revenues from the energy 'saved' as the result of the DR scheme.

Consumers: A combination of benefits from shifting to the lower 'night' tariff for heating and the additional discount offered by generators as a part of HSO DR scheme. The metrics here are the estimated pay-back period of the storage heaters and annual reduction in energy cost for the household after recouping the initial investment.

### Modelling Approach and Findings

Our modelling for this exercise was centered around options to shift a share of domestic demand for electricity. 

We began with a shift from 'regular' heating pattern to a schedule where consumers take maximum advantage of their new storage heaters and the opportunity to charge them at the 'off peak' rates. 

We saw, that if consumers had storage heaters and could move most of their heating demand to the 'off peak' hours, they could save £21/year on their annual electricity costs (or 5% of their default cost). If there was 50% or 80% discount offered on the 'saved' curtailment electricity, than the saving would increase to £49/year (12%) or £66/year (16%) respectively. The shift to 'off peak' hours also allowed to save 928 MWh of curtailed electricity.

```python
def shift_demand(df_consumed, midpoint=10):
    df_shifted = df_consumed.copy()

    df_shifted = df_shifted.reset_index()
    df_shifted["Timestamp"] -= df_shifted["Timestamp"].dt.hour.apply(
        lambda x: dt.timedelta(hours=int(x-(midpoint + np.sign(midpoint-x)*np.abs(midpoint-x)) % 24))
    )   
    df_shifted.set_index("Timestamp", inplace=True)
    if isinstance(df_consumed, pd.Series):
        return df_shifted[df_consumed.name]
    return df_shifted

fig, ax = plt.subplots(figsize=(10, 5))

actual = energy_consumed.groupby(energy_consumed.index.hour).mean()
shifted = shift_demand(energy_consumed)
shifted = shifted.groupby(shifted.index.hour).mean()

ax.plot(actual, label="Actual Consumption")
ax.plot(shifted, label="Shifted Consumption")
_ = ax.set_xlabel("Hour")
_ = ax.legend()
plt.title('Electric Heating Demand Shift to "Off-Peak" Hours')
plt.show()
```

After that we designed a shift of demand to match the curtailment profile and computed the reductions to consumer costs and curtailment reduction. 

As we expected, the reduction in consumer costs was more modest, as a large share of consultion remained in the higher 'day' tariff window. Under this shift the consumers could save £6/year on their annual electricity costs (or 1.5% of their default cost). If there was 50% or 80% discount offered on the 'saved' curtailment electricity, than the saving would increase to £30/year (7%) or £44/year (11%) respectively. The shift to 'off peak' hours also allowed to save 660 MWh of curtailed electricity.

### Further opportunities

The base case analysis above focuses on the Orkney residents who already use electric heating (4,465 out of 10,385 households). There is a potential to also extend the scheme to the residents relying on hydrocarbons for heating, if the terms could be made competitive with the current heating costs for these residents.

The main benefits of such conversion would be an expansion of the local demand for renewable-based electricity and reduction of the $CO_{2}$ emissions on the islands.

Based on Orkney statistical data for 2017<sup>8</sup>, 25% of residential space heating was fuled by oil and futher 14% - by solid fuel. (The remainder of the population had already converted to 'green' heat pumps and ground source heating.) 

As a part of our analysis we calculated that an average household which uses electric heating consumes 3,350 kWh of electricity per year, including 2,513 kWh on heating (75% of total). Using this number, we estimate that conversion of up to 39% of Orkney households to electric heating would increase Orkney electricity consumption by up to 10.2 GWh a year. 

If such conversion is to take place, it would be critical to ensure that the additional electricity demand is met by local RE generation (and not with additional import). To generate this amount of energy it would take 3 turbines of the same specifications as our sample turbine (0.9MW, operating with 45% capacity factor) to be fully dedicated to this incremental demand and the 100% DR flexibility to capture all generation.

The conversion from hydrocarbon-based to RE-based electricity would allow to reduce Orkney's hydrocarbon consumption and as the result cut the islands' $CO_{2}$ emissions by up to 4.6k tons a year<sup>14</sup>, which is equivalent to taking 1,000 passenger vehicles off the roads. 

In addition to the conversion of heating demand to electricity Orkney might consider introducing hydrogen or battery-based energy storage into the mix of the system balancing tool. That would allow not only to shift the consumption, but potentially take advantage of tariff arbitrage opportunities.

Based on the data analysis and the modelling we performed for this exercise we can conclude that there is definitely potential to reduce curtailment of wind generation in Orkney, reduce electricity costs for the islands' residents, and as the result - reduce fuel poverty levels.

Having said so, we can see that the cost of upgrding to modern storage heaters can be seen as prohibitive to a large share of the population, especially considering the relatively low level of income on the islands.

## Limitations of the analysis

Considering the limited amount of data available for this review, we made a number of large simplified assumptions, relating to operational and commercial aspects of the case, as well as to both the supply of and demand for electricity. 

If the decision to pursue the opportunity further is taken by Kaluza, a more refined set of assumptions will be required before any material commitments are made. That would include a broader review of power generation and consumption forecasts, as well as other local initiatives to increase efficiency of Orkney RE generation and the potential increase of export capacity. 

We based our analysis on the static historical data for one year, thus exposed to the hindsight bias. In reality, the Demand Response effort would need to rely on dynamic weather (wind), demand and price forecasting and would need to have immediate (automatic) response mechanisms embedded to it to achieve the optimal level of efficiency.
