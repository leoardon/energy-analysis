Telemetry from a single turbine on the Orkney Isles:

**Columns:**

- `Timestamp`: the datetime in UTC of the measurement (the timestamp is situated at the end of the measurement period)
- `Power_kw`: average power in kW during the measurement period
- `Setpoint_kw`: the mean maximum-allowed generator output during the measurement period
- `Wind_ms`: mean wind speed at the turbine during the measurement period

The setpoint of this turbine indicates whether the turbine is being forcibly curtailed by the network operator — the turbine controller is programmed to ensure that turbine output is always less than or equal to the setpoint. For example, if the setpoint is at the generator’s maximum output (900 kW), the generator can be assumed to be uncurtailed.

#### Power output as a function of wind speed

To estimate the relationship between wind speed and the power generated by the Orkney turbine, we use uncurtailed data, which we separate into 4 distinct segments:

- the low-wind segment (1), where the wind speed is insufficient for power generation;
- the ramp-up segment (2), with power output being a function of wind speed;
- full capacity segment (3), where power output is restricted by the turbine capacity;
- safety ramp-down segment (4), where the turbine operations are restricted for safety reasons due to high winds.

This separation is important, as the relationship between power and wind speed is driven by different factors in each of them. From the below analysis we will see that we can assume:

- 0kW power at wind speed up to 3m/s;
- increasing output, proportional to wind speed in the interval of 3-13 m/s;
- 900kW power (nameplate capacity) at the speed between 13 and 29 m/s;
- exponetially decreasing output beyond 29 m/s, depending on the wind, and untimate shutdown of the turbine.

In the EDA above we argued that zero output in the conditions of no curtailment and sufficient wind is likely to be an anomaly, therefore remove these records from the power curve analysis.

```python
telemetry_uc = telemetry[~telemetry['curtailed']].copy()
telemetry_uc = telemetry_uc.sort_values(by=["Wind_ms"])
```

```python
telemetry_uc.loc[
    (telemetry_uc["Wind_ms"] > 3) &
    (telemetry_uc["Wind_ms"] < 29) &
    (telemetry_uc["Power_kw"].abs() <= 1e-5)
] = np.nan
telemetry_uc.dropna(inplace=True)
```

```python
telemetry_uc_l = telemetry_uc[
    (telemetry_uc["Wind_ms"] > 3) &
    (telemetry_uc["Wind_ms"] < 29)
].copy()

#telemetry_uc_l.loc[telemetry_uc_l["Power_kw"].abs() <= 1e-5] = np.nan
#telemetry_uc_l.dropna(inplace=True)
```

```python
fig, ax = plt.subplots()
_ = telemetry_uc_l.plot(kind="scatter", x="Wind_ms", y="Power_kw", alpha = 0.01, ax=ax)
```

```python
from scipy.optimize import curve_fit

def logistic_func(x, k, x0):    
    return 900 / (1 + np.exp(-k * (x-x0)))

p0 = (1/80, 10)

popt_logis, _ = curve_fit(logistic_func, telemetry_uc_l["Wind_ms"], telemetry_uc_l["Power_kw"], p0)
```

```python
popt_logis
```

```python
y_hat = logistic_func(telemetry_uc_l["Wind_ms"], *popt_logis)
_ = telemetry_uc_l.plot(kind="scatter", x="Wind_ms", y="Power_kw", alpha = 0.01)
_ = plt.plot(telemetry_uc_l["Wind_ms"], y_hat, color="r")
```

```python
telemetry_uc_e = telemetry_uc[
    (telemetry_uc["Wind_ms"] >= 29)
]
```

```python
fig, ax = plt.subplots()
_ = telemetry_uc_e.plot(kind="scatter", x="Wind_ms", y="Power_kw", alpha = 0.01, ax=ax)
```

```python
def exp_func(x, a, b):    
    return a*(b**x)

p0 = (400e6, 1/2)

popt_exp, _ = curve_fit(exp_func, telemetry_uc_e["Wind_ms"], telemetry_uc_e["Power_kw"], p0)
```

```python
popt_exp
```

```python
y_hat = exp_func(telemetry_uc_e["Wind_ms"], *popt_exp)
_ = telemetry_uc_e.plot(kind="scatter", x="Wind_ms", y="Power_kw", alpha = 0.01)
plt.plot(telemetry_uc_e["Wind_ms"], y_hat, color="r")
```

```python
def power_uncurtailed(x):
    return np.where(
        x <= 3, 0, 
        np.where(
            x < 29, logistic_func(x, *popt_logis), 
            exp_func(x, *popt_exp)
        )
    )

y_hat = power_uncurtailed(telemetry_uc["Wind_ms"])
```

```python
_ = telemetry_uc.plot(kind="scatter", x="Wind_ms", y="Power_kw", alpha = 0.01)
plt.plot(telemetry_uc["Wind_ms"], y_hat, color="r")
```

#### Power loss due to curtailment

We use the previously fitted function to predict what would have been the power generated without curtailment and we evaluate the power loss.

```python
telemetry["Power_potential_kw"] = np.where(telemetry['curtailed'], 
                                           power_uncurtailed(telemetry["Wind_ms"]),
                                           telemetry['Power_kw'])
```

The estimated uncurtailed output allows us now to calculate the estimated losses of turbine output cased by curtailment.

```python
telemetry['est_loss_kw'] = telemetry['Power_potential_kw'] - telemetry['Power_kw']
m = telemetry["est_loss_kw"] < 0.
telemetry.loc[m, "est_loss_kw"] = 0.
```

#### Energy lost

We now transform that to energy. The data not being measured at a regular interval we interpolate the data in order to have the data for every minute. We will then be able to integrate the power over time to get the energy lost because of the curtailment.

```python
power_loss_df = telemetry.copy().drop(['curtailed'], axis = 1)

dt_min = power_loss_df.index.min().date()
dt_max = power_loss_df.index.max().date() + dt.timedelta(days=1)
period = (dt_max - dt_min).days * 24 * 60

t = power_loss_df.index
r = pd.date_range(dt_min, periods=period, freq='T')

power_loss_df_ri = power_loss_df.reindex(t.union(r)).interpolate('time').loc[r]
power_loss_df_ri.dropna(inplace=True)
```

We then compute the energy by taking the mean power generated in each hour.

```python
energy_hr = power_loss_df_ri.resample("h").mean()
energy_hr.rename(columns={
    'Power_kw':'energy_kwh', \
    'Power_potential_kw':'est_uncurtailed_kwh', \
    'est_loss_kw': 'est_loss_kwh'
}, inplace = True)
```

Looking at the distribution of the energy lost, we see that there is a lot of opportunity to save waste of energy. Fortunately, in the majority of the cases there is no loss, however we see that summing all the energy lost over the year comes to a big number in the end: 1099652 kwh or 15% of the uncurtailed output.

```python
_ = energy_hr[energy_hr['est_loss_kwh'] != 0]['est_loss_kwh'].hist(bins = 30, grid=False, figsize=(8, 4))
```

```python
energy_hr['est_loss_kwh'].sum()
```

**Demographic:** 
- We know that the number of households in 2017 was **10,385** <sup>3</sup>.

- We assume that **43%** of the population of Orkney Islands has electric heaters. For those who do, **75%** of the consumed relates to space heating <sup>4</sup>. As we are interested in the space heating consumption ('flexible' consumption which can be shifted in time), we identify and excude the 'other' electricity demand.


**Wind capacity:** 
- From our research we have found that the total wind capacity for the isles was **45,700kw**. Knowing that the data provided was for an average turbine of 900kw capacity we could compute the number of turbines present on the isles.


**Pricing:** 
- Generator: The Feed-in-Tariff paid by the government to promote renewable energy and received by the generators is estimated at **10p/kwh** generated. We also estimated based on our research that a generator could sell the energy at **3p/kwh**. (We will also include a variable simulating the price discount offered to the consumer in our model).
- Consumer: We have found from our research the day and night tariff for a consumer living in the Orkney Isles: **18.71p/kwh** and **13.15p/kwh**.
- Hardware: The device required to switch a consumer to the new system is composed of a smart unit with a cost of **£556** and heaters with a cost of **£162** a piece. We estimated that 1 smart heater per household would suffice.

### Demand shift

We have seen from the EDA of the demand data, that the consumption had 3 main regimes. With the introduction of a smart unit and storage heating, we can actually model a demand shift strategy where the smart unit will decide to consume energy during off-peak hours and stock it until it is needed. However, there is a caveat here: if all the households shift their consumption during off-peak hours, the off-peak hours will become peak hours and vice-versa. We leverage the fact that only 43% of the Orkney households are using electric heating.

*NB:* a more elaborated model could be develop to optimise the time at which the consumption timings taking into account the mass effect.

#### Shift to off-peak hours

```python
def shift_demand(df_consumed, midpoint=10):
    df_shifted = df_consumed.copy()

    df_shifted = df_shifted.reset_index()
    df_shifted["Timestamp"] -= df_shifted["Timestamp"].dt.hour.apply(
        lambda x: dt.timedelta(hours=int(x-(midpoint + np.sign(midpoint-x)*np.abs(midpoint-x)) % 24))
    )   
    df_shifted.set_index("Timestamp", inplace=True)
    if isinstance(df_consumed, pd.Series):
        return df_shifted[df_consumed.name]
    return df_shifted

fig, ax = plt.subplots(figsize=(10, 5))

actual = energy_consumed.groupby(energy_consumed.index.hour).mean()
shifted = shift_demand(energy_consumed)
shifted = shifted.groupby(shifted.index.hour).mean()

ax.plot(actual, label="Actual Consumption")
ax.plot(shifted, label="Shifted Consumption")
_ = ax.set_xlabel("Hour")
_ = ax.legend()
```

#### Shift to curtailed hours

A second strategy, which is the one we will be using for our simulation, is to use shift the demand to the period whith the most curtailed energy. That way we will be maximising the amount of energy saved from curtailment.

```python
SHIFT_HOUR_MAPPING = dict(
    zip(
        energy_consumed.groupby(energy_consumed.index.hour).mean().sort_values(by=["Energy_mean_kwh"]).index,
        energy_loss_df_2017.groupby(energy_loss_df_2017.index.hour).mean().sort_values(by=["Energy_loss_kwh"]).index
    )
)

def shift_demand_2(df_consumed):
    df_shifted = df_consumed.copy()

    df_shifted = df_shifted.reset_index()
    df_shifted["Timestamp"] -= df_shifted["Timestamp"].dt.hour.apply(
        lambda x: dt.timedelta(hours=x-SHIFT_HOUR_MAPPING[x])
    )
    df_shifted.set_index("Timestamp", inplace=True)
    if isinstance(df_consumed, pd.Series):
        return df_shifted[df_consumed.name]
    return df_shifted
```

### Energy  saved from curtailment

#### Energy Consumed

To compute the energy consumed, we use the average consumption of a household. In our model we parametrise the proportion of the people participating to the programme (`dr`), the total number of households in the islands (`n_households`), the proportion of households already using electric heating (`pop_elec_heating_prop`), the proportion of the consumption dedicated to electrical heating (`electric_heating_share`) and whether or not the smart system to shift the demand is enabled (`shift`).

```python
def compute_energy_consumed( #electricity consumed for heating under the scheme
    dr,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    shift=True
):
    """ Function to compute the amount of energy consumed if `dr`
    percent of the households participate to the programme
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_households (int): the total  number of households on the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        electric_heating_share (float): the proportion of consumption used for electric heating
        shift (bool): whether the demand should be shifted for the calculation
        
    Returns:
    --------
        energy_consumed (pd.Series): the amount of energy consumed
    """
    
    d = energy_consumed.copy().rename(columns={"Energy_mean_kwh": "Energy_consumed_kwh"})
    
    normal_demand = d["Energy_consumed_kwh"].copy()
    
    # the people taking part in the programme have smart heaters and have shifted their demand for heating only
    heating_demand = normal_demand * electric_heating_share
    if shift:
        heating_demand = shift_demand(heating_demand) #shift_demand_2
    other_demand = normal_demand * (1. - electric_heating_share)
    participant_demand = heating_demand + other_demand
    
    proportion_of_participants =  pop_elec_heating_prop * dr
    
    return n_households * (
        proportion_of_participants * participant_demand + 
        (1-proportion_of_participants) * normal_demand
    )
```

#### Energy Uncurtailed

We reuse the dataset cleaned up above to compute the energy lost. Because the data provided corresponds to a single turbine we pass the number of turbines available.

```python
def compute_energy_uncurtailed(n_turbines=NUMBER_TURBINES):
    d = energy_uncurtailed_df_2017.copy().rename(columns={"Energy_kwh": "Energy_uncurtailed_kwh"})
    return d["Energy_uncurtailed_kwh"] * n_turbines

def compute_energy_produced(n_turbines=NUMBER_TURBINES):
    d = energy_produced_df_2017.copy().rename(columns={"Energy_kwh": "Energy_produced_kwh"})
    return d["Energy_produced_kwh"] * n_turbines
```

#### Energy saved

```python
def compute_energy_saved_from_curtailment(
    dr,
    n_turbines=NUMBER_TURBINES,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    shift=True
):
    """ Function to compute the amount of energy saved from curtailment if `dr`
    percent of the households participate to the programme
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_turbines (int): the number of  turbines available for power generation
        n_households (int): the total  number of households on the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        electric_heating_share (float): the proportion of consumption used for electric heating
        shift (bool): whether the demand should be shifted for the calculation
        
    Returns:
    --------
        saved_from_curtailment (pd.Series): the amount of energy saved
    """
    
    consumed = compute_energy_consumed( 
        dr,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        shift=shift
    )
    produced = compute_energy_produced(n_turbines)
    
    return (consumed + energy_impexp - produced).clip(0)
```

### Money Saved

#### Customers

The programme benefits the customers in multiple ways. Firstly, thanks to the smart unit, the demand could partially shift towards off-peak hours, which have a better rate. Secondly, the programme can  offer a customer discount in order to attract customer and because this energy would have been lost otherwise. We evaluate now the impact of joining the programme for a customer.

```python
def compute_money_paid_by_customers(#electricity cost to consumers under the scheme
    dr,
    n_turbines=NUMBER_TURBINES,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    day_price=E7_DAY_TARIFF,
    night_price=E7_NIGHT_TARIFF,
    cust_discount=PRICE_DISCOUNT,
    shift=True
):
    """ Function to compute the amount of paid
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_turbines (int): the number of  turbines available for power generation
        n_households (int): the total  number of households on the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        electric_heating_share (float): the proportion of consumption used for electric heating
        day_price (float): the price of the electricity during the day
        night_price (float): the price of the electricity during the night
        cust_discount (float): the price discount applied on the energy saved from curtailment consumed
        shift (bool): whether the demand should be shifted for the calculation
        
    Returns:
    --------
        money_paid_df (pd.Series): the money paid by the customers
    """
    energy_consumed_df = compute_energy_consumed(
        dr,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        shift=shift
    )
    
    energy_saved_from_curtailment_df = compute_energy_saved_from_curtailment(
        dr,
        n_turbines=n_turbines,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        shift=shift
    )
    
    price_f = lambda h: customer_price(h, night_price=night_price, day_price=day_price)
    price = pd.Series(energy_consumed_df.index.hour, index=energy_consumed_df.index).apply(price_f)
    
    # Apply discount only on uncurtailed energy
    return (
        (energy_consumed_df - energy_saved_from_curtailment_df) * price +
        energy_saved_from_curtailment_df * price * (1-cust_discount)
    ).rename("Paid_by_customers")
```

```python
def compute_money_saved_by_customers(#electricity cost to consumers under the scheme
    dr,
    n_turbines=NUMBER_TURBINES,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    day_price=E7_DAY_TARIFF,
    night_price=E7_NIGHT_TARIFF,
    cust_discount=PRICE_DISCOUNT,
    shift=True
):
    """ Function to compute the amount of money saved
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_turbines (int): the number of  turbines available for power generation
        n_households (int): the total  number of households on the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        electric_heating_share (float): the proportion of consumption used for electric heating
        day_price (float): the price of the electricity during the day
        night_price (float): the price of the electricity during the night
        cust_discount (float): the price discount applied on the energy saved from curtailment consumed
        shift (bool): whether the demand should be shifted for the calculation
        
    Returns:
    --------
        money_saved_df (pd.Series): money saved by the customers
    """
    return (compute_money_paid_by_customers(
        0,
        n_turbines=n_turbines,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        day_price=day_price,
        night_price=night_price,
        cust_discount=0,
        shift=False
    ) - compute_money_paid_by_customers(
        dr,
        n_turbines=n_turbines,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        day_price=day_price,
        night_price=night_price,
        cust_discount=cust_discount,
        shift=shift
    )).rename("Saved_by_customers").round(2)
```

```python
money_saved_shift_only = compute_money_saved_by_customers(
    DR,
    cust_discount=0.,
    shift=True
).sum()
money_saved_shift_only_c =  money_saved_shift_only / (DR * NUMBER_HOUSEHOLDS * POP_ELEC_HEATING_PROP)

money_saved_shift_and_discount = compute_money_saved_by_customers(
    DR,
    cust_discount=0.5,
    shift=True
).sum()
money_saved_shift_and_discount_c = money_saved_shift_and_discount / (DR * NUMBER_HOUSEHOLDS * POP_ELEC_HEATING_PROP)

display(Markdown((
    f"If {DR:.0%} of the households join the programme "
    f"simply by switching to smart heating, the average yearly amount saved per customer "
    f"would be **£{money_saved_shift_only_c:,.2f}**, if we also add a discount of 50%, the average yearly amount saved "
    f"would be **£{money_saved_shift_and_discount_c:,.2f}**."
)))
```

#### Generators

```python
def compute_money_saved_generators(#saved revenues for generators
    dr, 
    n_turbines=NUMBER_TURBINES,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    subsidy=FiT,
    tariff=EXPORT_TARIFF,
    cust_discount=PRICE_DISCOUNT,
    shift=True
):
    """ Function to compute the amount of money saved
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_turbines (int): the number of  turbines available for power generation
        n_households (int): the total  number of households on the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        electric_heating_share (float): the proportion of consumption used for electric heating
        subsidy (float): the Feed-in-Tariff received by the generator
        tariff (float): the tariff of the electricty
        cust_discount (float): the price discount applied on the energy saved from curtailment consumed
        shift (bool): whether the demand should be shifted for the calculation
        
    Returns:
    --------
        money_saved_df (pd.Series): money saved by generators
    """
    energy_saved_df = compute_energy_saved_from_curtailment(
        dr, 
        n_turbines=n_turbines,
        n_households=n_households,
        pop_elec_heating_prop=pop_elec_heating_prop,
        electric_heating_share=electric_heating_share,
        shift=shift
    )
    
    price = generator_tariff(subsidy, tariff)
        
    return (energy_saved_df * price * (1-cust_discount)).rename("Saved_by_generators").round(2)
```

### Installation costs

```python
def compute_installation_costs(
    dr,
    n_households=NUMBER_HOUSEHOLDS,
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    n_heaters=N_HEATERS,
    heater_cost=HEATER_COST,
    smart_unit_cost=SMART_UNIT_COST
):
    """ Function to compute the installation costs
    
    Params:
    -------
        dr (float): percentage of households participating to the programme
        n_households (int): the number of households in the islands
        pop_elec_heating_prop (float): the proportion of households with electric  heating
        n_heaters (int): the average number of heaters per household
        heater_cost (float): the cost for one heater
        smart_unit_cost (float): the cost for a smart unit
        
    Returns:
    --------
        installation_costs (float): the amount of money spent to install the new system
        
    Notes:
    ------
        We assume that the costs include the salary of the engineer doing the installation
    """
    return dr * n_households * pop_elec_heating_prop * (n_heaters * heater_cost + smart_unit_cost)
```

### Plot

```python
def profit_curve( # generators
    n_households=NUMBER_HOUSEHOLDS,
    n_turbines=NUMBER_TURBINES,
    
    pop_elec_heating_prop=POP_ELEC_HEATING_PROP,
    electric_heating_share=ELEC_HEATING_SHARE_CONS,
    
    n_heaters=N_HEATERS,
    heater_cost=HEATER_COST,
    heater_cost_gen_share=HEATER_COST_GEN_SHARE,
    smart_unit_cost=SMART_UNIT_COST,
    smart_unit_cost_gen_share=SMART_UNIT_COST_GEN_SHARE,
    
    subsidy = FiT,
    tariff = EXPORT_TARIFF,
    
    day_price=E7_DAY_TARIFF,
    night_price=E7_NIGHT_TARIFF,
    price_discount=PRICE_DISCOUNT*100,
    shift=True
):
    price_discount /= 100
    x = np.linspace(0, 1, 101)
    
    money_saved_by_customers = []
    insallation_costs_customers = []
    money_saved_by_generators = []
    insallation_costs_generators = []
    
    for i in x:
        money_saved_by_customers.append(
            compute_money_saved_by_customers(
                i,
                n_turbines=n_turbines,
                n_households=n_households,
                pop_elec_heating_prop=pop_elec_heating_prop,
                electric_heating_share=electric_heating_share,
                day_price=day_price,
                night_price=night_price,
                cust_discount=price_discount,
                shift=shift
            ).sum()
        )
        insallation_costs_customers.append(
            compute_installation_costs(
                i,
                n_households=n_households,
                pop_elec_heating_prop=pop_elec_heating_prop,
                n_heaters=n_heaters,
                heater_cost=heater_cost*(1-heater_cost_gen_share),
                smart_unit_cost=smart_unit_cost*(1-smart_unit_cost_gen_share)
            )
        )
        
        money_saved_by_generators.append(
            compute_money_saved_generators(
                i,
                n_turbines=n_turbines,
                n_households=n_households,
                pop_elec_heating_prop=pop_elec_heating_prop,
                electric_heating_share=electric_heating_share,
                subsidy=subsidy,
                tariff=tariff,
                cust_discount=price_discount,
                shift=shift,
            ).sum()
        )
        insallation_costs_generators.append(
            compute_installation_costs(
                i,
                n_households=n_households,
                pop_elec_heating_prop=pop_elec_heating_prop,
                n_heaters=n_heaters,
                heater_cost=heater_cost*heater_cost_gen_share,
                smart_unit_cost=smart_unit_cost*smart_unit_cost_gen_share
            )
        )

    profit = np.array(money_saved_by_generators)\
            - np.array(money_saved_by_customers) - np.array(insallation_costs_generators)
    customer_saved = np.array(money_saved_by_customers) - np.array(insallation_costs_customers)
    
    display(Markdown(f"Generator Max Profit:**£{profit.max():,.2f}** for {profit.argmax():}% DR"))    
    
    fig, ax = plt.subplots(1, figsize=(15, 4))
    
    _ = ax.set_title("Generator Profit")
    _ = ax.set_xlabel("DR")
    _ = ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    _ = ax.plot(x, profit, c='k')
    _ = ax.axvline(x=x[profit.argmax()], ymax=profit.max(), linestyle='--', c='b')
    _ = ax.axhline(xmax=x[profit.argmax()], y=profit.max(), linestyle='--', c='b')
    _ = ax.scatter(x[profit.argmax()], profit.max(), s=50, c='b')
    _ = ax.fill_between(x, 0, profit, where=profit >= 0, color="C2", alpha=0.5)
    _ = ax.fill_between(x, 0, profit, where=profit <= 0, color="C3", alpha=0.5)
    
```
